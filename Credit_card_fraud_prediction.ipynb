```python
#i have to change the class from float to int before i can use it for the target
credit_card_data.Class=credit_card_data['Class'].astype(int)
```

```python
# as can be seen fromabove, the data is skewed towards the legit transactions
# i cant feed this data into the ML algo because we have very very little case of fruadulent data
#less than 1% when cmpared to each other, highly unbalanced data
# 0--> normal transaction
# 1--> fraud transaction
```

```python
# now we want to separate the data for more analysis 
legit = credit_card_data[credit_card_data.Class == 0]
fraud = credit_card_data[credit_card_data.Class == 1]
```

```python
#you can see there is quite a different between the mean of the legit and the fraud
#that is going to hel with the classification

#because of the massive size different of the two data types,we now want to do something
#somthing called undersampling 
```

**Under-Sampling**
we want to build a new dataset from the legit sample that is identical to the fraud sample

fraud sample --> 56
so we have to built a random sample of 56 rows from the the legit

```python
legit_sample = legit.sample(n=56)
```

**Concatinating two dataframes**

now we have to concatinate the new legit sample and the fraud sample

```python
new_dataset = pd.concat([legit_sample,fraud], axis=0)
```

```python
new_dataset.head()
```

```python
new_dataset.tail()
```

```python
new_dataset.shape
```

```python
new_dataset.Class.value_counts()
```

```python
new_dataset.groupby('Class').mean()
```

```python
#even with the new samller= dataset, the difference btw the two is still quite a lot
# and across board 
#it tells us that the nature of the dataset hasnt changed so can still be used accuractely 

#if the samples were bad, we would have had very different results but we dont in this case 
```

Splitting the data 

```python
X = new_dataset.drop('Class',axis=1)
Y= new_dataset.Class
```

```python
X.head()
```

```python
Y.head()
```

splitting to training and test data

```python
X_train,X_test,Y_train,Y_test = train_test_split(X,Y,test_size=0.1,random_state=2,stratify=Y)
#stratify is for the even distribution of the 1s and zeros 
```

```python
print(X_train.shape,Y_train.shape)
```

Logistics Regression

```python
model = LogisticRegression(max_iter=1000)
```

```python
model.fit(X_train,Y_train)

# #putting the max_iter shoild be the last resort.
#  try the other things thta can b done first
#  standardization etc
```

```python
#now to predict the accuracy score of our model 
```

```python
training_data_prediction = model.predict(X_train)
```

```python
training_data_prediction_accuracy = accuracy_score(Y_train,training_data_prediction)
print('the accuracy score of our training data prediction is :', training_data_prediction_accuracy)
```

```python
test_data_prediction = model.predict(X_test)
```

```python
test_data_prediction_accuracy = accuracy_score(Y_test,test_data_prediction)
print('the test data accuracy is:', test_data_prediction_accuracy)
```
